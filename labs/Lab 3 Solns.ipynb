## Problem 3

```python
#include <iostream>

using namespace std;

int main() {
    const int NICKES_TO_PENNIES = 5;
    const int DIMES_TO_PENNIES = 10;
    const int QUARTER_TO_PENNIES = 25;
    const int DOLLARS_TO_PENNIES = 100;

    int pennies = 0, nickles = 0, dimes = 0, quarters = 0, dollars = 0;

    cout << "How many pennies do you have?" << endl;
    cin >> pennies;

    cout << "How many nickles do you have?" << endl;
    cin >> nickles;

    cout << "How many dimes do you have?" << endl;
    cin >> dimes;

    cout << "How many quarters do you have?" << endl;
    cin >> quarters;

    cout << "How many dollars do you have?" << endl;
    cin >> dollars;

    int total = 0;
    total = pennies +
            nickles * NICKES_TO_PENNIES +
            dimes * DIMES_TO_PENNIES +
            quarters * QUARTER_TO_PENNIES +
            dollars * DOLLARS_TO_PENNIES;
    cout << "The value of all of your coins is " << total;
    return 0;
}
```
